```python
import pandas as pd
def convert_DEATH_cell(cell):       #here if i want to change any value in death i can use def function here you can see in death in assam row first it was 2.0 after we changed it converted to yes
    if cell==2.0:
        return 'yes'
    return cell
df=pd.read_excel("C:\\Users\Kavali Charani\Documents\excel\india_covid.xlsx","india_covid",converters={
    'DEATH':convert_DEATH_cell
})
df
```

### if i want to write to df in one excel file in two diff sheets

```python
df_name=pd.DataFrame({
    'first':['PINKY','SONY','MEGHU'],
    'mark':[24,25,20],
    'grade':[1,1,1],
    'max':[5,5,4]
             
  })
df_person=pd.DataFrame({
    'last':['KAVALI','BK','ANCHALA'],
    'place':['atp','atp','atp'],
    'pe':[1,2,3]
})
```

```python
with pd.ExcelWriter('name_person.xlsx')as writer:   #here we created two dataframes and merged in single excel file
    df_name.to_excel(writer,sheet_name='name')
    df_person.to_excel(writer,sheet_name='person')
```
